Min Max Normalization

```python
cols=['crim','zn','tax','black']
for col in cols:
    minimum=min(df_house[col])
    maximum=max(df_house[col])
    df_house[col]=(df_house[col]-minimum)/(maximum-minimum)
```

```python
#Standardization
from sklearn import preprocessing
scaler=preprocessing.StandardScaler()
#fit our data
scaled_cols=scaler.fit_transform(df_house[cols])
scaled_cols=pd.DataFrame(scaled_cols,columns=cols)
scaled_cols.head()
```

```python
#Train Test Split
```

```python
X=df_house.drop(columns=['medv','rad'],axis=1)
Y=df_house['medv']
```

```python
print(X.shape)
print(Y.shape)
```

```python
#model Training
```

```python
from sklearn.model_selection import cross_val_score,train_test_split
from sklearn.metrics import mean_squared_error
def train(model,x,y):
    #train the model
    X_train,X_test,Y_train,Y_test=train_test_split(x,y,random_state=42)
    model.fit(X_train,Y_train)
    #predict the training set
    pred=model.predict(X_test)
    #perform cross validation
    cv_score=cross_val_score(model,x,y,scoring='neg_mean_squared_error',cv=5)
    cv_score=np.abs(np.mean(cv_score))
    
    print('model report')
    print('mean squared error:',mean_squared_error(Y_test,pred))
    print('cv score:',cv_score)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression(normalize=True)
train(model,X,Y)
coef=pd.Series(model.coef_,X.columns).sort_values()
coef.plot(kind='bar',title='Model Coeff')
```

```python
from sklearn.tree import DecisionTreeRegressor
model=DecisionTreeRegressor()
train(model,X,Y)
coef=pd.Series(model.feature_importances_,X.columns).sort_values(ascending=False)
coef.plot(kind='bar',title='Model Coeff')
```

```python
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor()
train(model,X,Y)
coef=pd.Series(model.feature_importances_,X.columns).sort_values(ascending=False)
coef.plot(kind='bar',title='Model Coeff')
```
